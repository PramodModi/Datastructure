```python
class LRU:
    def __init__(self, maxSize):
        self.maxSize = maxSize or 1
        self.cache = {}
        self.size = 0
        self.mostRecentUsed = DoublyLinkedList
    def insertKeyValuePair(self, key, value):
        #check if key is already there in cache
        # If yes, then update the value
        # If not then then check the size
            # if size is equal to maxsize then remove the tail from mostrecent used
        #add the new node in cache
        # Update the mostrecentused set the head
        pass
    def getValueFromKey(self, key):
        # check if key is there in the cache
        # if no then return None
        # else get the value, update the mostrecentused, set the head then return the value
        pass
    def mostRecentKey(self):
        # return the key of mostrecent used head
        pass
    

    
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    def setHead(self, node):
        if self.head is None:
            self.head = node
            self.tail = node
            return
        if self.head == self.tail:
            self.head.prev= node
            node.next = self.head
            self.head = node
            self.tail.prev = node
            return
        if self.tail == node:
            self.tail = self.tail.prev
        node.removeBinding()
        node.next = self.head
        self.head.prev = node
        self.head = node
    
    def removeTail(self):
        if self.tail is None:
            return
        if self.tail == self.head:
            self.head = None
            self.tail = None
            return
        self.tail = self.tail.prev
        self.tail = None
               
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None
        self.prev = None
    
    def removeBinding(self):
        if self.next is not None:
            self.next.prev = self.prev
        if self.prev is not None:
            self.prev.next = self.next
        self.next = None
        self.prev = None
```
