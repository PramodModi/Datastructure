```python
import queue
```

```python
class Empty(Exception):
    pass
```

```python
# Stack class
class Stack:
    def __init__(self):
        self._data = []
        self._size = 0
    
    def isEmpty(self):
        return self._size == 0
    
    def push(self, e):
        self._data.append(e)
        self._size += 1
        
    def pop(self):
        if self.isEmpty():
            raise Empty("Stack is empty")
        e = self._data.pop()
        self._size -= 1
        return e
    
    def top(self):
        if self.isEmpty():
            raise Empty("Stack is empty")
        return self._data[-1]
    
    def size(self):
        return self._size
            
```

```python
# Binary Tree class
class BinaryTree:
    # Node class for binary tree
    class _Node:
        # Slots is being used to save memory and faster attribute access
        __slots__ = '_data', '_left', '_right'
        
        def __init__(self, data):
            self._data = data
            self._left = None
            self._right = None
    
    # Init class for BinaryTree
    def __init__(self):
        self._root = None
        
    def root(self):
        return self._root
    
    def insert(self, data):
        newNode = self._Node(data)
        
        if self._root == None:
            self._root = newNode
            return
        Q = queue.Queue(maxsize = 20)
        Q.put(self._root)
        while(Q.empty() is not True):
            tmp = Q.get()
            if tmp._left:
                Q.put(tmp._left)
            else:
                tmp._left = newNode
                Q.queue.clear()
                return
            if tmp._right:
                Q.put(tmp._right)
            else:
                tmp._right = newNode
                Q.queue.clear()
                return
            
        Q.queue.clear()
    
    # Traversing
    def preorder(self):
        if self._root == None:
            raise Empty("Binary Tree is empty")
        
        S= Stack()
        S.push(self._root)
        while (S.isEmpty() is not True):
            tmp = S.pop()
            print(tmp._data)
            if tmp._right:
                S.push(tmp._right)
            if tmp._left:
                S.push(tmp._left)
  
    def inorder(self):
        if self._root == None:
            raise Empty("Binary Tree is empty")

        S= Stack()
        tmp = self._root
        while(True): 
            while tmp:
                S.push(tmp)
                tmp = tmp._left
            if S.isEmpty():
                break;
            tmp = S.pop()
            print(tmp._data)
            tmp = tmp._right
            
    def postorder(self):
        if self._root == None:
            raise Empty("Binary Tree is empty")
        S = Stack()
        tmp = self._root
        element = None
        while True:
            while tmp:
                S.push(tmp)
                tmp = tmp._left
            if S.isEmpty():
                break
            tmp = S.pop()
            if tmp._right and tmp._right != element:
                S.push(tmp)
                tmp = tmp._right
            else:
                element = tmp
                print(element._data)
                tmp = None
           
    def levelorder(self):
        if self._root is None:
            raise Empty("Tree is empty")
        Q = queue.Queue(maxsize = 20)
        Q.put(self._root)
        while (Q.empty() is not True):
            temp = Q.get()
            print(temp._data)
            if temp._left:
                Q.put(temp._left)
            if temp._right:
                Q.put(temp._right)
        Q.queue.clear()     
            
```
