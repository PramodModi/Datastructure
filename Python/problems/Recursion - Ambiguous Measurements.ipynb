## Problem:

This problem deals with measuring cups that are missing important measuring labels. Specifically, a
measuring cup only has two measuring lines, a Low (L) line and a High (H) line. This means that these
cups can't precisely measure and can only guarantee that the substance poured into them will be
between the L and H line. 

For example, you might have a measuring cup that has a Low line at 400ml
and a high line at 435ml . This means that when you use this measuring cup, you can only be sure
that what you're measuring is between 400ml and 435ml .

Note that:
Each measuring cup will be represented by a pair of positive integers [L, H] , where
0 <= L <= H .

will always satisfy the following constraint: 0 <= low <= high .


Once you've measured some liquid, it will immediately be transferred to a larger bowl that will
eventually (possibly) contain the target measurement.
You can't pour the contents of one measuring cup into another cup.


Example:
