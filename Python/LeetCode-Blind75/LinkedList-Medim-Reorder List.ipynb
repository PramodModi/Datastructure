## 143. Reorder List

You are given the head of a singly linked-list. The list can be represented as:

L0 → L1 → … → Ln - 1 → Ln

Reorder the list to be on the following form:

L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …

You may not modify the values in the list's nodes. Only nodes themselves may be changed.

 

Example 1:


Input: head = [1,2,3,4]

Output: [1,4,2,3]

Example 2:


Input: head = [1,2,3,4,5]

Output: [1,5,2,4,3]
 

Constraints:

The number of nodes in the list is in the range [1, 5 * 104].

1 <= Node.val <= 1000

```python
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

def reorderList(head):
        """
        Do not return anything, modify head in-place instead.
        """
        if head.next is None or head.next.next is None:
            return
        
        firstHalfHead = head
        
        secondHalfHead = splitList(head)
        secondHalfReversed  = reverseList(secondHalfHead)
        finalHead = interweaveList(firstHalfHead, secondHalfReversed)
        head = finalHead

def splitList(head):
    cur = head
    temp  = head
    while temp and temp.next:
        cur = cur.next
        temp = temp.next.next
    splitHead = cur.next
    cur.next = None
    return splitHead
    
def reverseList(head):
    if head.next is None:
        return head
    prev = head
    cur = head.next
    temp = None
    while cur:
        temp = cur.next
        cur.next = prev
        prev = cur
        cur = temp
    head.next = None
    head = prev
    return head
def interweaveList(head1, head2):
    first = head1
    second = head2
    while first and second:
        temp = first.next
        first.next = second
        first = temp
        temp = second.next
        second.next = first
        second = temp
    return head1
    
```
