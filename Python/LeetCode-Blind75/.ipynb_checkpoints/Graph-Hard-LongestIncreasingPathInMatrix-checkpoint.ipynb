## 329. Longest Increasing Path in a Matrix
Given an m x n integers matrix, return the length of the longest increasing path in matrix.

From each cell, you can either move in four directions: left, right, up, or down. You may not move diagonally or move outside the boundary (i.e., wrap-around is not allowed).

 

Example 1:

Input: matrix = [[9,9,4],[6,6,8],[2,1,1]]

Output: 4

Explanation: The longest increasing path is [1, 2, 6, 9]

Example 2:

Input: matrix = [[3,4,5],[3,2,6],[2,2,1]]

Output: 4

Explanation: The longest increasing path is [3, 4, 5, 6]. Moving diagonally is not allowed.

Example 3:

Input: matrix = [[1]]

Output: 1
 

Constraints:

m == matrix.length
n == matrix[i].length
1 <= m, n <= 200
0 <= matrix[i][j] <= 231 - 1

```python
def longestIncreasingPath(matrix):
        m = len(matrix)
        n = len(matrix[0])
        
        maxLength = 0
        dp = [[-1 for _ in range(n)] for _ in range(m)]
        
        for i in range(m):
            for j in range(n):
                maxLength = max(maxLength, traverse(i,j,matrix, dp))
        return maxLength
    
def traverse(r, c, matrix, dp):
    if dp[r][c] != -1:
        return dp[r][c]
    
    count = 1
    
    ngbrs = getNeighbours(r,c,matrix)
    for row, col in ngbrs:
        if matrix[row][col] > matrix[r][c]:
            count = max(count, 1 + traverse(row, col, matrix,dp))
        
    dp[r][c] = max(dp[r][c], count)
    return dp[r][c]

def getNeighbours(r,c,matrix):
    ngbrs = []
    if r > 0:
        ngbrs.append((r-1, c))
    if r < len(matrix) - 1:
        ngbrs.append((r+1, c))
    if c > 0:
        ngbrs.append((r, c-1))
        
    if c < len(matrix[0]) - 1:
        ngbrs.append((r, c+1))
        
    return ngbrs
        
```

```python
matrix = [[9,9,4],[6,6,8],[2,1,1]]
longestIncreasingPath(matrix)
```

```python
matrix = [[3,4,5],[3,2,6],[2,2,1]]
longestIncreasingPath(matrix)
```
