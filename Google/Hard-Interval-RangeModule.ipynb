## Problem: Range Module
LeetCode : 715
https://leetcode.com/problems/range-module/description/

A Range Module is a module that tracks ranges of numbers. Design a data structure to track the ranges represented as half-open intervals and query about them.

A half-open interval [left, right) denotes all the real numbers x where left <= x < right.

Implement the RangeModule class:

    RangeModule() Initializes the object of the data structure.
    void addRange(int left, int right) Adds the half-open interval [left, right), tracking every real number in that interval. Adding an interval that partially overlaps with currently tracked numbers should add any numbers in the interval [left, right) that are not already tracked.
    boolean queryRange(int left, int right) Returns true if every real number in the interval [left, right) is currently being tracked, and false otherwise.
    void removeRange(int left, int right) Stops tracking every real number currently being tracked in the half-open interval [left, right).
     

Example 1:

    Input
    ["RangeModule", "addRange", "removeRange", "queryRange", "queryRange", "queryRange"]
    [[], [10, 20], [14, 16], [10, 14], [13, 15], [16, 17]]
    Output
    [null, null, null, true, false, true]
    
    Explanation
    RangeModule rangeModule = new RangeModule();
    rangeModule.addRange(10, 20);
    rangeModule.removeRange(14, 16);
    rangeModule.queryRange(10, 14); // return True,(Every number in [10, 14) is being tracked)
    rangeModule.queryRange(13, 15); // return False,(Numbers like 14, 14.03, 14.17 in [13, 15) are not being tracked)
    rangeModule.queryRange(16, 17); // return True, (The number 16 in [16, 17) is still being tracked, despite the remove operation)
 

Constraints:

    1 <= left < right <= 109
    At most 104 calls will be made to addRange, queryRange, and removeRange.


```python
class RangeModule:

    def __init__(self):
       self.intervals = []
        

    def addRange(self, left, right):
        # Binary search, find the interval, having start value should be greatest but smaller than left. 
        # Insert this new interval at the next position
        if len(self.intervals) == 0 or self.intervals[-1][1] <= left:
            self.intervals.append([left, right])
        else:
            index = binarySearch(self.intervals, left);
            print(index)
            #self.intervals = self.intervals[0:index] + [[left, right]] + self.intervals[index:]
            self.intervals.insert(index, [left, right])
            print(self.intervals)

        # Merge interval
        res = [self.intervals[0]]
        for i in range(1, len(self.intervals)):
            if res[-1][1] >= self.intervals[i][0]:
                res[-1][1] = max(res[-1][1], self.intervals[i][1])
            else:
                res.append(self.intervals[i])
        self.intervals = res
        print(self.intervals)
            
    
    def queryRange(self, left, right):
        # DO binary search and find the index, where intervals[0] is maximum but lesser than or equal to left.
        idx = binarySearch(self.intervals, left)
        print(idx)
        if not self.intervals:
            return False
        if self.intervals[idx][0] > left: idx -= 1
        return self.intervals[idx][1] >= right
        

    def removeRange(self, left, right):
        
        # After deleting Keep new ranges in temp list and at the end assign back to self.intervals.
        temp = []
        # 4 conditions.
        # 1. When remove interval contains all intervals. left <= intervals[i][0] and right >= intervals[i][1], delete this interval
        # 2. When remove interval is outside of range. left >= intervals[i][1] or right <= intervals[i][0] , No need to delete this interval
        # 3. Overlapping condition, when either end of remove interval is outside of range
            #1. When left is smaller than interva;[i][0], then after deleting we will have [right, inrevals[i][1]]
            #2. when right is bigger than interval[i][1], then after deleting we will have [interval[i][0], left]
        #4. When remove interval is inside of the range. Then break interval in two parts and store.

        for i in range(len(self.intervals)):
            interval = self.intervals[i]
            if left <=  interval[0] and right >= interval[1]:
                continue
            elif left >= interval[1] or right <= interval[0]:
                temp.append(interval)
            elif left < interval[0]:
                temp.append([right, interval[1]])
            elif right > interval[1]:
                temp.append([interval[0], left])
            else:
                temp.append([interval[0], left])
                temp.append([right, interval[1]])
        self.intervals = temp
        print(self.intervals)

def binarySearch(array, left):
    
    l, r = 0, len(array)-1
    while l <= r:
        mid = (l+r)//2
        if left < array[mid][0]:
            r = mid-1
        elif left > array[mid][0]:
            l = mid+1
        else:
            l =  mid
            break
    return l
        
        
```

```python
obj = RangeModule()
obj.addRange(10,20)
obj.removeRange(14, 16)
param_2 = obj.queryRange(10,14)
print(param_2)
param_2 = obj.queryRange(13,15)
print(param_2)
param_2 = obj.queryRange(16,17)
print(param_2)
```
