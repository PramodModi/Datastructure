## Problem : N-Queens

LeetCode: 51

https://leetcode.com/problems/n-queens/description/

The n-queens puzzle is the problem of placing n queens on an n x n chessboard such that no two queens attack each other.

Given an integer n, return all distinct solutions to the n-queens puzzle. You may return the answer in any order.

Each solution contains a distinct board configuration of the n-queens' placement, where 'Q' and '.' both indicate a queen and an empty space, respectively.

 

Example 1:



    Input: n = 4
    Output: [[".Q..","...Q","Q...","..Q."],["..Q.","Q...","...Q",".Q.."]]
    Explanation: There exist two distinct solutions to the 4-queens puzzle as shown above
Example 2:

    Input: n = 1
    Output: [["Q"]]
 

Constraints:

    1 <= n <= 9

```python
def solveNQueens(n):
        board = [['.'] * n for _ in range(n)]
        result = []
        solve(board, 0, result)
        return result
def solve(board, row, result):
    if row == len(board):
        result.append(["".join(board[r]) for r in range(len(board))])
        return
    for col in range(len(board[row])):
        if isAllowed(board, row, col):
            board[row][col] = 'Q'
            solve(board, row+1, result)
            board[row][col] = '.'
def isAllowed(board, row, col):
    for r in range(row):
        if 'Q' == board[r][col]:
            return False
        if (col-row+r)>= 0 and board[r][col-row+r] == 'Q':
            return False
        if (col+row-r) < len(board[0]) and board[r][col+row-r] == 'Q':
            return False
    return True
```
